```python
import ast
```

```python
# Function to fetch only genre names from genre columns

def Get_Names(text):
    temp = []
    for i in (ast.literal_eval(text)):
        temp.append(i['name'])
    return temp
```

```python
df['genres'] = df['genres'].apply(Get_Names)
```

```python
df.head()
```

```python
# Same function will be used with keywords and cast columns

df.iloc[0].keywords
```

```python
df['keywords'] = df['keywords'].apply(Get_Names)
```

```python
df.head()
```

```python
df.iloc[0].cast
```

```python
def Fetch_Cast(text):
    names = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter<3:
            names.append(i['name'])
            counter+=1
        else:
            break
    return names
```

```python
df['cast'] = df['cast'].apply(Fetch_Cast)
df.head(2)
```

```python
df.iloc[0].crew
```

```python
def Fetch_Cast(text):
    names = []
    for i in ast.literal_eval(text):
        if (i['job'] == 'Director' or i['job'] == 'Producer'):
            names.append(i['name'])
    return names
```

```python
df['crew'] = df['crew'].apply(Fetch_Cast)
df.head(2)
```

```python
df.iloc[0].crew
```

```python
df.iloc[1].crew
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
df.iloc[0].overview
```

```python
df.head(2)
```

```python
def RemoveSpaces(text):
    temp = []
    for i in text:
        temp.append(i.replace(" ",""))
    return temp
```

```python
df['genres'] = df['genres'].apply(RemoveSpaces)
df['keywords'] = df['keywords'].apply(RemoveSpaces)
df['cast'] = df['cast'].apply(RemoveSpaces)
df['crew'] = df['crew'].apply(RemoveSpaces)
```

```python
df.head(2)
```

```python
df['Features'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.head(2)
```

```python
final_data = df.drop(['genres','keywords','overview','cast','crew'],axis=1)
final_data.head()
```

```python
final_data['Features'] = final_data['Features'].apply(lambda x:" ".join(x))
```

```python
final_data.head()
```

```python
final_data['Features'] = final_data['Features'].apply(lambda x:x.lower())
final_data.head()
```

```python
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
```

```python
print(final_data['Features'].apply(lambda x:x.split())[0])
```

```python
def stem(text):
    temp = []
    for i in text.split():
        temp.append(porter.stem(i))
    return " ".join(temp)
```

```python
final_data['Features'] = final_data['Features'].apply(stem)
final_data.head(2)
```

```python
final_data.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vector = cv.fit_transform(final_data['Features']).toarray()
```

```python
vector[0]
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
final_data[final_data['title'] == 'Avatar'].index
```

```python
def Recommend(movie):
    index = final_data[final_data['title'] == movie].index[0]
    distances = similarity[index]
    
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    
# movies_list mein movie ke index,uske similarity stored hai isliye hum movies[0] krke us particular similarity vali movie
# ka index nikalenge fir use apne dataframe mein iloc krke dalenge taaki uska title pata chal ske   

    for movies in movies_list:
        print(final_data.iloc[movies[0]].title)
```

```python
Recommend('Aliens')
```

```python
Recommend('Avatar')
```

```python
Recommend('Spider-Man')
```

```python
Recommend('The Dark Knight Rises')
```
